8.4.1 [**Q- what are the sales for these top 20 premium cars?**](#Section8)<br>

```python
def get_totalcarsales(r):
    return data[(data.car == r.car) & (data.price == r.max_price)].car.count()

def get_latestyearofsale(r):
    return data[(data.car == r.car) & (data.price == r.max_price)].year.max()
```

```python

cars_with_max_price_df['total_sales'] = cars_with_max_price_df.apply(get_totalcarsales, axis=1)
cars_with_max_price_df['recently sold on'] = cars_with_max_price_df.apply(get_latestyearofsale, axis=1)
```

```python
f, axs = plt.subplots(figsize=(15,15))
sns.scatterplot(data=cars_with_max_price_df.sort_values('max_price', ascending=False).head(20) ,x='recently sold on', y='max_price',hue= 'car')
```

 8.6 [**Q - What are the best mileage cars in each segment of 'class'?**](#Section8.6)<br>

```python
def classify_class(r):
    if r.price <= 10000:
        return "Economy"
    elif (r.price > 10000) & (r.price <= 30000):
        return "Luxury"
    else:
        return "Premium"
```

```python

# Set new column called 'class' for defining the sement
import pandas as pd
data = pd.read_csv(filepath_or_buffer = 'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/car_sales.csv', encoding='cp1252')
data["class"] = data.apply(classify_class, axis=1)
```

```python

best_cars = data.pivot_table(index=['class','car'],  values="mileage", aggfunc='max')
best_cars.reset_index(inplace=True)
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(data=best_cars, x='class', y='mileage', palette='autumn')

plt.title('Mileage vs Class')
```
